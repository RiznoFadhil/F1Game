# Modeling

## F1 2021 

### Linear Regression 

```python
attributes = ['AWA', 'EXP', 'RAC', 'PAC']
x21 = f1_21[attributes]
y21 = f1_21['RTG']

x21_train, x21_test, y21_train, y21_test = train_test_split(x21, y21, test_size=0.2, random_state=0)
```

```python
lm = LinearRegression()
model_21 = lm.fit(x21_train, y21_train)

y21_pred = lm.predict(x21_test)

# Calculate the Mean Squared Error (MSE) and R^2 score
mse21 = mean_squared_error(y21_test, y21_pred)
r2_21 = r2_score(y21_test, y21_pred)

mse21,r2_21
```

```python
model_21.intercept_
```

```python
model_21.coef_
```
